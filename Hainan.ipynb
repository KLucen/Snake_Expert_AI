```python
# 根据条件筛选数据
filter_data = df[df['生态学特征Ecological traits_地理分布（中国）Geographic distribution（China）'].str.contains('海南', na=False)]

# 查看筛选后的数据
filter_data
```

```python
# 保存筛选后的数据到新的 Excel 文件
filter_data.to_excel('海南蛇类特征数据集_海南蛇类.xlsx', index=False)
```

```python
# 筛选有毒蛇的数据
venomous_snakes = filter_data[filter_data['生态学特征Ecological traits_有无毒性 Venomous or not'] == '有毒venomous']

# 筛选无毒蛇的数据
non_venomous_snakes = filter_data[filter_data['生态学特征Ecological traits_有无毒性 Venomous or not'] == '无毒non-venomous']

# 保存有毒蛇数据到 Excel 文件
venomous_snakes.to_excel('海南有毒蛇类.xlsx', index=False)

# 保存无毒蛇数据到 Excel 文件
non_venomous_snakes.to_excel('海南无毒蛇类.xlsx', index=False)
```
